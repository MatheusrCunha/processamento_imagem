```python
from ultralytics import YOLO

model = YOLO('yolov8n.pt')
```

```python
results = model('moedas_comluz.jpeg')
```

```python
for result in results:
  boxes = result.boxes
  result.save(filename='moedas_comluz.jpeg.jpg')
```

```python
import matplotlib.pyplot as plt
import matplotlib.image as mpimg

img = '/content/moedas_comluz.jpeg'

image = mpimg.imread(img)
plt.imshow(image)
plt.axis('off')
plt.show()
```
